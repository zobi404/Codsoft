# Features Selection

```python
X = df[['V1', 'V3', 'V4', 'V7', 'V9', 'V10', 'V11', 'V12', 'V14', 'V16', 'V17', 'V18']]
Y = df['Class']
X.shape
```

# Imbalanced Dataset Problem

## Performed Undersampling

```python
nm = NearMiss()
```

```python
x, y = nm.fit_resample(X, Y)
```

```python
print(x.shape, y.shape)
```

```python
print('Before Sampling {}'.format(Counter(Y)))
print('After Sampling {}'.format(Counter(y)))
```

# Splitting Dataset into Training and Testing Set

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

# Model Selection

```python
model = LogisticRegression(solver = 'liblinear')
```

```python
model
```

```python
model.fit(x_train, y_train)
```

## Making Predictions

```python
pred = model.predict(x_test)
pred
```

# Model Evaluation

```python
acc=accuracy_score(y_test, pred)
```

```python
f1=f1_score(y_test, pred)
```

```python
prob = model.predict_proba(x_test)
```

```python
loss = log_loss(pred,prob)
```

## Final Report

```python
score = {
    'Model' : ['Logistic Regression'],
    'Accuracy_Score' : [f'{round(acc*100, 2)}%'],
    'F1 Score' : [round(f1*100, 2)],
    'Log Loss' : [f'{round(loss*100, 2)}%']
}

metrics = pd.DataFrame(data = score)
metrics
```
